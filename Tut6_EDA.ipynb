```python
# import necessary libraries
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv("C:/Users/kh/Documents/WQD7001/products_order_by_months.csv")
df1
```

```python
df1["Months"] = df1["Months"].replace({"April": "Apr", "June": "Jun", "July": "Jul"})
```

```python
# Create a line plot
fig, ax = plt.subplots()
for product, group in df1.groupby('Products'):
    group.plot(x='Months', y='Number_of_Orders', ax=ax, label=product)

# Add a title and axis labels
ax.set_title("Number of Orders by Product and Month")
ax.set_xlabel("Month")
ax.set_ylabel("Number of Orders")

# Display the plot
plt.show()
```

From line graph above, Beauty Cream has the lowest number of order compared to other products. Skin Lightening Cream has the highest order number but it is not stable.

```python
df2 = pd.read_csv("C:/Users/kh/Documents/WQD7001/sales_by_category.csv")
df2
```

```python
# Define a colormap
colors = plt.cm.get_cmap('Set3')
```

```python
# Create a bar chart
fig, ax = plt.subplots()
df2.plot(kind='bar', x='Products', y='Sales', ax=ax, color=colors(range(len(df2))))

# Add a title and axis labels
ax.set_title("Sales by Product")
ax.set_xlabel("Product")
ax.set_ylabel("Sales")

# Display the plot
plt.show()
```

From the above bar chart, Lip gloss is the higest sales compared to the other products while Rouge is the least sales.
